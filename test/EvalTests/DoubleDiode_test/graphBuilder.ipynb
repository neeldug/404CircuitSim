```python
import pandas as pd
import matplotlib.pyplot as plot
import matplotlib.cm as cm
import numpy as np
import matplotlib.colors as colors

a = pd.read_csv('converge1.68s.csv')
#a.plot(x='vGuess[0]',y='vGuess[1]')

x = a['vGuess[0]'].to_numpy()
y = a['vGuess[1]'].to_numpy()
c = a['Step'].to_numpy()
error = a['Magnitude of Error /V'].to_numpy()
fig, axs = plot.subplots(1,2,figsize=(20,10))
plot.subplots_adjust(wspace=0.1)
errormap=colors.LinearSegmentedColormap.from_list('gr',["g", "r"], N=256)
itmap=colors.LinearSegmentedColormap.from_list('rg',["cyan", "navy"], N=256)
axs[1].scatter(x=x,y=y,c=error,cmap=errormap,marker='x')
axs[1].set_title("Guessed Voltage across D2 /V vs Guessed Voltage across D1 /V\n With the D2 in forward conduction\nColor Gradient - Error Magnitude")
axs[0].scatter(x=x,y=y,c=c,cmap=itmap,marker='x')
axs[0].minorticks_on()
axs[1].minorticks_on()
axs[1].grid(True,which="major")
axs[1].grid(True,which="minor",  linestyle=':',linewidth=0.5)
axs[0].grid(True,which="major")
axs[0].grid(True,which="minor",  linestyle=':',linewidth=0.5)
axs[1].set(xlabel="Voltage Across of D1 /V")
axs[1].set(ylabel="Voltage Across of D2 /V")
axs[0].set(xlabel="Voltage Across of D1 /V")
axs[0].set(ylabel="Voltage Across of D2 /V")
axs[0].set(title="Guessed Voltage across D2 /V vs Guessed Voltage across D1 /V\n With the D2 in forward conduction\nColor Gradient - Iteration Step")
fig.colorbar(cm.ScalarMappable(cmap=errormap), ax=axs[1], orientation='horizontal')
fig.colorbar(cm.ScalarMappable(cmap=itmap), ax=axs[0], orientation='horizontal')
fig.savefig("graphs/ConvergenceDiagram.png", bbox_inches='tight')
```
